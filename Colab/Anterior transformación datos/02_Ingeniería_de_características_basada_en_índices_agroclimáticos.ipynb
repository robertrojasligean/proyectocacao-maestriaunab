## ESTRÉS TÉRMICO MENSUAL (ET)
Con el fin de cuantificar el impacto de la temperatura máxima sobre el cultivo de cacao, se construyó la variable Estrés Térmico Mensual (ET), la cual mide el grado de desviación de la temperatura máxima diaria respecto al valor óptimo para el desarrollo fisiológico del cultivo. Para su cálculo, se consideró como temperatura óptima central un valor de 28 °C, con un rango de tolerancia entre 24 °C y 32 °C, de acuerdo con la literatura agroclimática del cacao.

A nivel diario, el estrés térmico se estimó como la desviación absoluta de la temperatura máxima diaria (TMAX) respecto al valor óptimo, normalizada en una escala entre 0 y 1. Posteriormente, esta variable diaria fue agregada a escala mensual mediante el promedio de los valores diarios, obteniendo así el Estrés Térmico Mensual (ET).

En el conjunto de datos final, esta variable se almacena en la columna denominada ET, indexada por el período year_month, y toma valores continuos entre 0 (condiciones térmicas óptimas) y 1 (alto nivel de estrés térmico). Esta variable constituye uno de los componentes fundamentales del Índice de Estrés Agroclimático Mensual del Cacao (IEAM-C) y se utiliza como insumo para el posterior modelado con técnicas de machine learning.

ET ∈ [0, 1]

0 = condición térmica óptima

1 = alta desviación respecto al óptimo térmico del cacao

El Estrés Térmico Mensual (ET) es una variable normalizada que toma valores entre 0 y 1, donde valores bajos indican condiciones térmicas cercanas al óptimo para el cultivo de cacao y valores altos representan una mayor desviación respecto a dicho óptimo. Para facilitar su interpretación agronómica, el ET se clasifica en los siguientes rangos:





| **Rango ET** | **Nivel de estrés térmico** | **Interpretación agronómica**                                               |
| ------------ | --------------------------- | --------------------------------------------------------------------------- |
| 0.00 – 0.20  | Muy bajo                    | Condiciones térmicas cercanas al óptimo para el desarrollo del cacao        |
| 0.21 – 0.40  | Bajo                        | Leve desviación térmica, sin impacto significativo en el cultivo            |
| 0.41 – 0.60  | Moderado                    | Estrés térmico perceptible, con posible afectación de procesos fisiológicos |
| 0.61 – 0.80  | Alto                        | Estrés térmico elevado, con riesgo para floración y llenado de mazorca      |
| 0.81 – 1.00  | Muy alto                    | Condiciones térmicas adversas, con alto riesgo agronómico                   |


Estos rangos permiten clasificar los meses según su nivel de estrés térmico y facilitan el análisis comparativo temporal, así como la integración de esta variable dentro del Índice de Estrés Agroclimático Mensual del Cacao (IEAM-C).

| Expresión                            |
| ------------------------------------ |
| **ETₘ = ( Σᵈ I(T_d > T_opt) ) / Dₘ** |


| Símbolo   | Descripción                                               |
| --------- | --------------------------------------------------------- |
| **ETₘ**   | Índice de Estrés Térmico del mes *m*                      |
| **Dₘ**    | Número total de días del mes *m*                          |
| **T_d**   | Temperatura diaria considerada                            |
| **T_opt** | Umbral térmico óptimo para el cultivo de cacao            |
| **I(·)**  | Función indicadora: 1 si T_d > T_opt, 0 en caso contrario |





```python
df['estres_termico_dia'] = (
    abs(df['TMAX'] - 28) / 8
).clip(0, 1)

```

```python
df.columns
```

```python
# Construcción del Estrés Térmico Mensual (ET)
df_ET_mensual = (
    df
    .groupby("year_month")
    .agg(
        dias_mes=("estres_termico_dia", "count"),
        dias_estres=("estres_termico_dia", "sum")
    )
    .reset_index()
)

# Cálculo del índice ET
df_ET_mensual["ET"] = df_ET_mensual["dias_estres"] / df_ET_mensual["dias_mes"]

# Visualización inicial
df_ET_mensual.head()
```

## Estrés Hídrico Mensual (EH)

El Estrés Hídrico Mensual (EH) es un índice agroclimático que representa la proporción de días dentro de un mes en los que la disponibilidad hídrica resulta insuficiente para el desarrollo adecuado del cultivo de cacao. Este indicador permite identificar periodos con potencial limitación hídrica, los cuales pueden afectar procesos fisiológicos clave del cultivo, como la floración, el cuajado de frutos y el llenado de mazorcas.

En el presente estudio, el estrés hídrico diario se define como una variable binaria, en la cual un día es clasificado como estresante cuando la precipitación diaria es inferior a un umbral mínimo de 1 mm. Este valor es comúnmente utilizado en estudios agroclimáticos en regiones tropicales para identificar días secos efectivos. A partir de esta variable diaria, el índice mensual se calcula como la razón entre el número de días con estrés hídrico y el total de días del mes correspondiente.

El Estrés Hídrico Mensual se calcula mediante la siguiente expresión:

EH_m = (Número de días con estrés hídrico en el mes) / (Número total de días del mes)

El índice EH toma valores entre 0 y 1, donde valores cercanos a 0 indican condiciones hídricas favorables y valores cercanos a 1 reflejan una alta frecuencia de días con déficit hídrico. Este indicador no tiene como objetivo la predicción de condiciones climáticas futuras, sino la caracterización y clasificación del comportamiento agroclimático histórico mensual, sirviendo como insumo para la generación de escenarios de apoyo a la planificación agrícola del cultivo de cacao.

```python
# =====================================================
# Normalización de nombres de columnas
# =====================================================

df.columns = (
    df.columns
      .str.lower()
      .str.strip()
      .str.replace(' ', '_')
      .str.normalize('NFKD')
      .str.encode('ascii', errors='ignore')
      .str.decode('utf-8')
)

df.columns
```

```python
# ============================================================
# CÁLCULO DEL ESTRÉS HÍDRICO (EH)
# Nivel diario → agregación mensual → integración al modelo
# ============================================================

# ------------------------------------------------------------
# 1. Estrés hídrico diario
# Un día presenta estrés hídrico si la precipitación < 1 mm
# ------------------------------------------------------------

df['estres_hidrico_dia'] = (df['precipitacion'] < 1).astype(int)

# ------------------------------------------------------------
# 2. Agregación mensual
# Se calcula la proporción de días con estrés hídrico por mes
# ------------------------------------------------------------

eh_mensual = (
    df.groupby('year_month')
      .agg(
          EH=('estres_hidrico_dia', 'mean')
      )
      .reset_index()
)

# ------------------------------------------------------------
# 3. Integración del EH mensual en el DataFrame del modelo
# ------------------------------------------------------------

df_mensual = df_mensual.merge(
    eh_mensual,
    on='year_month',
    how='left'
)

# ------------------------------------------------------------
# 4. Verificación rápida
# ------------------------------------------------------------

df_mensual[['year_month', 'ET', 'EH']].head()
df_mensual[['ET', 'EH']].describe()
```

```python
# ============================================================
# LIMPIEZA FINAL DEL DATAFRAME MENSUAL
# Eliminar variables intermedias de conteo
# ============================================================

columnas_intermedias = [
    'dias_mes',
    'dias_estres'
]

df_mensual = df_mensual.drop(columns=columnas_intermedias, errors='ignore')

# Verificación final del modelo
df_mensual.head()
df_mensual.describe()
```

## El Estrés por Radiación Solar Mensual (ER)
se define como la proporción de días dentro de un mes en los que la radiación solar normalizada alcanza valores elevados, representando condiciones potencialmente adversas para el cultivo de cacao. Dado que el cacao es una especie adaptada a ambientes de sombra parcial, la ocurrencia recurrente de altos niveles de radiación puede generar estrés fisiológico, incrementar la demanda hídrica y afectar negativamente el desarrollo del cultivo.

En este estudio, la radiación solar se aborda como una variable normalizada, por lo que el estrés radiativo se establece mediante un umbral relativo, considerando como día con estrés por radiación aquel en el que la radiación solar normalizada es mayor o igual a 0.75. El índice ER se calcula a nivel mensual como el promedio de la variable diaria binaria, lo cual equivale a la proporción de días con radiación elevada respecto al total de días del mes. Finalmente, el índice se clasifica en niveles de riesgo bajo, moderado y alto, con el fin de apoyar la planificación agrícola del cultivo de cacao.

#Rangos

| ER mensual       | Nivel de riesgo | Interpretación                                 |
| ---------------- | --------------- | ---------------------------------------------- |
| ER < 0.30        | Bajo            | Condiciones de radiación mayormente favorables |
| 0.30 ≤ ER < 0.60 | Moderado        | Exposición frecuente a radiación elevada       |
| ER ≥ 0.60        | Alto            | Estrés radiativo severo                        |


# Valores optimos

| Variable                    | Valor / rango de referencia | Interpretación agronómica        |
| --------------------------- | --------------------------- | -------------------------------- |
| Tipo de cultivo             | Sotobosque                  | Requiere sombra parcial          |
| Exposición solar óptima     | Moderada                    | Favorece fotosíntesis sin estrés |
| Radiación solar elevada     | Percentil ≥ 75              | Riesgo de estrés radiativo       |
| Radiación solar normalizada | 0 – 1                       | Variable relativa, no absoluta   |


```python
# ============================================================
# CORRECCIÓN DEL ESTRÉS POR RADIACIÓN SOLAR (ER)
# Umbral basado en percentil histórico
# ============================================================

# Calcular umbral estadístico (percentil 75)
umbral_radiacion = df['rad_sol'].quantile(0.75)

# Estrés por radiación diario (definición corregida)
df['estres_radiacion_dia'] = (df['rad_sol'] >= umbral_radiacion).astype(int)

# Agregación mensual
er_mensual = (
    df.groupby('year_month')
      .agg(
          ER=('estres_radiacion_dia', 'mean')
      )
      .reset_index()
)

# Eliminar ER previo si existe (evitar duplicados)
df_mensual = df_mensual.drop(columns=['ER'], errors='ignore')

# Integrar ER corregido al modelo
df_mensual = df_mensual.merge(
    er_mensual,
    on='year_month',
    how='left'
)

# Verificación
df_mensual[['ET', 'EH', 'ER']].describe()
```

## Estrés por Humedad Relativa Mensual (EHR)

La humedad relativa constituye una variable agroclimática relevante para el cultivo de cacao, debido a su influencia directa sobre los procesos fisiológicos de transpiración, balance hídrico de la planta y susceptibilidad a estrés combinado con altas temperaturas. De acuerdo con la literatura agronómica, el cacao presenta un desarrollo óptimo bajo condiciones de humedad relativa promedio comprendidas entre el 70 % y el 90 %, rangos que favorecen la actividad fotosintética y reducen el estrés hídrico atmosférico.

En este estudio, se definió el Estrés por Humedad Relativa diario como la condición en la cual la humedad relativa registrada se encuentra por debajo del umbral mínimo óptimo del 70 %, considerándose este valor como un límite inferior crítico para el cultivo. Bajo este criterio, cada día fue clasificado de forma binaria como día con estrés por humedad relativa o día sin estrés.

A partir de esta clasificación diaria, se construyó el Índice de Estrés por Humedad Relativa Mensual (EHR), definido como la proporción de días del mes que presentaron condiciones de estrés por humedad relativa. Este índice se expresa en una escala continua entre 0 y 1, donde valores cercanos a 0 indican meses con condiciones de humedad mayoritariamente favorables, mientras que valores cercanos a 1 representan meses con una alta recurrencia de condiciones de humedad relativa por debajo del rango óptimo para el cultivo de cacao.

Con el fin de facilitar la interpretación de los resultados, los valores del EHR fueron tipificados en tres rangos de riesgo agroclimático mensual: bajo, intermedio y alto. Se consideraron meses con bajo riesgo hídrico-atmosférico aquellos con valores de EHR inferiores a 0.40, riesgo intermedio los meses con valores entre 0.40 y 0.65, y riesgo alto aquellos con valores iguales o superiores a 0.65. Esta tipificación permite integrar el componente de humedad relativa dentro del análisis agroclimático mensual y fortalecer la evaluación del estrés ambiental que enfrenta el cultivo de cacao.

```python
# ============================================================
# CONSTRUCCIÓN DEL ESTRÉS POR HUMEDAD RELATIVA (EHR)
# ============================================================

# --- Verificación de columnas necesarias ---
columnas_hr = ['hr', 'year_month']
for col in columnas_hr:
    if col not in df.columns:
        raise ValueError(f"La columna '{col}' no existe en el DataFrame diario")

# ============================================================
# 1️⃣ CÁLCULO DIARIO DE ESTRÉS POR HUMEDAD RELATIVA
# ============================================================

# Umbral mínimo óptimo para cacao (%)
HR_MIN_OPTIMO = 70

# Variable binaria diaria de estrés por humedad relativa
df['estres_hr_dia'] = (df['hr'] < HR_MIN_OPTIMO).astype(int)

# ============================================================
# 2️⃣ AGREGACIÓN MENSUAL DEL EHR
# ============================================================

ehr_mensual = (
    df.groupby('year_month')
      .agg(
          dias_mes=('estres_hr_dia', 'count'),
          dias_estres_hr=('estres_hr_dia', 'sum')
      )
      .reset_index()
)

# Cálculo del índice EHR mensual
ehr_mensual['EHR'] = ehr_mensual['dias_estres_hr'] / ehr_mensual['dias_mes']

# ============================================================
# 3️⃣ INTEGRACIÓN DEL EHR EN df_mensual
# ============================================================

df_mensual = df_mensual.merge(
    ehr_mensual[['year_month', 'EHR']],
    on='year_month',
    how='left'
)

# ============================================================
# 4️⃣ VALIDACIÓN RÁPIDA
# ============================================================

df_mensual[['EHR']].describe()
```

## Construcción de variable Objetivo

El Índice de Estrés Agroclimático Mensual del Cacao (IEAM-C) se definió como la variable objetivo del proyecto y se construyó mediante la integración de los principales factores climáticos que influyen en el desarrollo fisiológico y productivo del cultivo de cacao: el estrés térmico (ET), el estrés hídrico (EH), el estrés por radiación solar (ER) y el estrés por humedad relativa (EHR). Cada uno de estos componentes fue previamente calculado a escala mensual y normalizado en un rango entre 0 y 1, donde valores cercanos a 0 indican condiciones favorables y valores cercanos a 1 representan escenarios de mayor nivel de estrés.

La formulación del IEAM-C se realizó como una combinación ponderada de los índices parciales, asignando las ponderaciones en función de la relevancia agronómica de cada variable y de su recurrencia histórica en el área de estudio. En este sentido, el componente hídrico recibió el mayor peso debido a su impacto directo en la disponibilidad de agua para el cultivo, seguido por el estrés térmico y el estrés por radiación solar. El estrés por humedad relativa fue incorporado con una ponderación menor, dado que el análisis histórico evidenció una baja frecuencia de condiciones críticas asociadas a esta variable. Matemáticamente, el índice se expresa como:

Este índice integrado permite sintetizar el nivel de estrés agroclimático mensual y constituye una herramienta de apoyo a la planificación agrícola del cultivo de cacao, sin pretender realizar predicciones determinísticas del comportamiento climático futuro.

# Rangos


| IEAM-C               | Escenario  | Interpretación agroclimática                   |
| -------------------- | ---------- | ---------------------------------------------- |
| IEAM-C < 0.40        | Favorable  | Condiciones mayormente óptimas para el cultivo |
| 0.40 ≤ IEAM-C < 0.65 | Intermedio | Presencia de estrés agroclimático moderado     |
| IEAM-C ≥ 0.65        | Crítico    | Alto nivel de estrés agroclimático             |



# Impacto

| Componente | Peso     | Justificación                               |
| ---------- | -------- | ------------------------------------------- |
| EH         | **0.45** | Principal limitante del cacao               |
| ET         | **0.30** | Impacta floración y cuajado                 |
| ER         | **0.20** | Afecta fotosíntesis                         |
| EHR        | **0.05** | Favorece o limita enfermedades y fisiología |



```python
# =========================================
# CONSTRUCCIÓN DE LA VARIABLE OBJETIVO IEAM-C
# =========================================

# Ponderaciones definidas según relevancia agronómica
w_EH  = 0.45   # Estrés hídrico
w_ET  = 0.30   # Estrés térmico
w_ER  = 0.20   # Estrés por radiación solar
w_EHR = 0.05   # Estrés por humedad relativa

# Cálculo del Índice de Estrés Agroclimático Mensual del Cacao
df_mensual['IEAM_C'] = (
    w_EH  * df_mensual['EH'] +
    w_ET  * df_mensual['ET'] +
    w_ER  * df_mensual['ER'] +
    w_EHR * df_mensual['EHR']
)

# =========================================
# CLASIFICACIÓN DE ESCENARIOS MENSUALES
# =========================================

def clasificar_ieam(valor):
    if valor < 0.45:
        return 'Favorable'
    elif valor < 0.60:
        return 'Intermedio'
    else:
        return 'Crítico'

df_mensual['Escenario_IEAM_C'] = df_mensual['IEAM_C'].apply(clasificar_ieam)

# =========================================
# ORDEN FINAL DE COLUMNAS
# =========================================

orden_columnas = [
    'year_month',
    'ET',
    'EH',
    'ER',
    'EHR',
    'IEAM_C',
    'Escenario_IEAM_C'
]

df_mensual = df_mensual[orden_columnas]

# Verificación
df_mensual.head()
```
