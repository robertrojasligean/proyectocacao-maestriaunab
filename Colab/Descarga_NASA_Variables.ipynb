```python
# ===============================
# 1. LIBRERÍAS
# ===============================
import requests
import pandas as pd
from datetime import datetime

# ===============================
# 2. PARÁMETROS DEL MUNICIPIO
# ===============================
LAT = 6.881
LON = -73.414

START_DATE = "20100101"
END_DATE   = "20241231"

# Variable NASA POWER
PARAMETER = "ALLSKY_SFC_SW_DWN"  # Radiación solar diaria (kWh/m2/día)

# ===============================
# 3. URL NASA POWER
# ===============================
url = (
    "https://power.larc.nasa.gov/api/temporal/daily/point"
    f"?parameters={PARAMETER}"
    f"&community=AG"
    f"&longitude={LON}"
    f"&latitude={LAT}"
    f"&start={START_DATE}"
    f"&end={END_DATE}"
    f"&format=JSON"
)

# ===============================
# 4. DESCARGA DE DATOS
# ===============================
response = requests.get(url)
data = response.json()

# ===============================
# 5. CONVERTIR A DATAFRAME
# ===============================
radiation_dict = data["properties"]["parameter"][PARAMETER]

df_radiation = pd.DataFrame(
    list(radiation_dict.items()),
    columns=["fecha", "radiacion_solar"]
)

# ===============================
# 6. FORMATEAR FECHA
# ===============================
df_radiation["fecha"] = pd.to_datetime(df_radiation["fecha"], format="%Y%m%d")

# ===============================
# 7. ORDENAR Y REVISAR
# ===============================
df_radiation = df_radiation.sort_values("fecha").reset_index(drop=True)

df_radiation.head(), df_radiation.tail()
```

```python
# Guardar a CSV
df_radiation.to_csv(
    "radiacion_solar_nasa_san_vicente_chucuri_2010_2024.csv",
    index=False
)
```

```python
# Guardar a Excel
df_radiation.to_excel(
    "radiacion_solar_nasa_san_vicente_chucuri_2010_2024.xlsx",
    index=False
)
```

```python
# =========================================================
# DESCARGA TEMPERATURA MAX Y MIN - NASA POWER
# SAN VICENTE DE CHUCURÍ (2010–2024)
# =========================================================

# 1. LIBRERÍAS
import requests
import pandas as pd
from google.colab import files

# 2. COORDENADAS Y FECHAS
LAT = 6.881
LON = -73.414

START_DATE = "20100101"
END_DATE   = "20241231"

# 3. VARIABLES NASA POWER
PARAMETERS = ["T2M_MAX", "T2M_MIN"]

# 4. URL NASA POWER
url = (
    "https://power.larc.nasa.gov/api/temporal/daily/point"
    f"?parameters={','.join(PARAMETERS)}"
    f"&community=AG"
    f"&longitude={LON}"
    f"&latitude={LAT}"
    f"&start={START_DATE}"
    f"&end={END_DATE}"
    f"&format=JSON"
)

# 5. DESCARGA
response = requests.get(url)
data = response.json()

# 6. EXTRAER DATOS
tmax_dict = data["properties"]["parameter"]["T2M_MAX"]
tmin_dict = data["properties"]["parameter"]["T2M_MIN"]

# 7. CREAR DATAFRAME
fechas = pd.to_datetime(
    list(tmax_dict.keys()),
    format="%Y%m%d"
)

df_temp = pd.DataFrame({
    "fecha": fechas,
    "temp_max": list(tmax_dict.values()),
    "temp_min": list(tmin_dict.values())
})

df_temp = df_temp.sort_values("fecha").reset_index(drop=True)

# 8. VALIDACIÓN RÁPIDA
print("Filas:", len(df_temp))
df_temp.head(), df_temp.tail()

# 9. GUARDAR ARCHIVOS
df_temp.to_csv(
    "temperatura_nasa_san_vicente_chucuri_2010_2024.csv",
    index=False
)

df_temp.to_excel(
    "temperatura_nasa_san_vicente_chucuri_2010_2024.xlsx",
    index=False
)

# 10. DESCARGAR ARCHIVOS
files.download("temperatura_nasa_san_vicente_chucuri_2010_2024.csv")
files.download("temperatura_nasa_san_vicente_chucuri_2010_2024.xlsx")
```
